```python
x = range(-10, 10)
y1 = [3 * d**2  for d in x]
y2 = [5 * d**3 + 2 *d **2  for d in x]
y3 = [d ** 3 for d in x]
plt.plot(x, y1, marker="o", linestyle="dashed", color="green", label="Y1")
plt.plot(x, y2, marker="*", linestyle="solid", color="red", label="Y2")
plt.plot(x, y3, marker="+", linestyle="dashed", color="m", label="Y3")
plt.title("functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper center")
plt.show()
```

```python
x1, y1 = range(0,5), [61, 45, 32, 13, 35]
x2, y2 = range(0,5), [17, 39, 46, 40, 27]
labels = ["A", "B", "C", "D", "E"]
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.bar(x1, y1, color="b", tick_label=labels)
ax1.set_title("dog")
ymin, ymax = plt.ylim()
ax2 = fig.add_subplot(1,2,2)
ax2.bar(x2, y2, color="g", tick_label=labels)
ax2.set_title("cat")
plt.ylim(ymin, ymax)
plt.show()
```

```python
x1, y1 = range(0,5), [61, 45, 32, 13, 35]
x2, y2 = range(0,5), [17, 39, 46, 40, 27]
labels = ["A", "B", "C", "D", "E"]
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True)
ax1.bar(x1, y1, color="b", tick_label=labels)
ax1.set_title("dog")
ax2.bar(x2, y2, color="g", tick_label=labels)
ax2.set_title("cat")
plt.show()
```
