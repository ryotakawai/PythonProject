```python
from random import randint
numbers = []
while len(numbers) < 10:
    n = randint(-10,100)
    if n < 0:
        print("中断されました")
        break
    if n in numbers:
        continue
    numbers.append(n)
else:
    print(numbers)
```

```python
numlist = {3, 4.2, 10, "apple", 1, 9}
sum = 0
for num in numlist:
    if not isinstance(num, (int, float)):
        print(num, "数値ではありません")
        break
    sum += num
    print(num, "/", sum)
```
